## Data cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"call.xlsx")
df
```

```python
## Row 19 and 20 are duplicate rows so we drop them with this code

df = df.drop_duplicates()
df
```

```python
#drop the not useful column

df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# working on last name column
#to strip unwanted values on the left and right side, lstrip for left, rstrip for right
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")  
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")

#But is is better to do it this way 1 shot, 123 for nums 
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
#CLeaning phone numbers

#This means replacing anything that isnt a to z / 0-9 with ' ' aka nothing 
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)


#Changing each value into a str using lambda func -> basically a lambda is a much quicker for loop
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))


#Before you do reformat with lambda, you have to change entire column to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])

#replacing na-- and Na-- with nothing
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na-', '')

df
```

```python
#Cleaning Address
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', n=2,  expand=True)
df
```

```python
#changing yes and no to same format from paying customer and do not contact

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')


#to also make sure all NA or Nan cells are replaced with nothing
df = df.replace('N/a', '')
df = df.fillna('')
df
```

## Once we are done with cleaning the data, now we have to find the customers that are actually interested to be called


```python
#remove rows from do not contact if they tell us not to contact
for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == 'Y':
        df.drop(i, inplace = True)
df
```

```python
#remove rows from phone number if they dont have phone number
for i in df.index:
    if df.loc[i, "Phone_Number"] == '' or df.loc[i, "Phone_Number"] == '-':
        df.drop(i, inplace = True)
df
```

```python
#to reset index since its all messed up after cleaning
df = df.reset_index(drop = True)
```
